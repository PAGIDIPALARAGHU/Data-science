```python
import pandas as pd
data=pd.read_csv('obesity.csv')
data.head(50)
obesity_level=data['NObeyesdad'].unique()
print(obesity_level)
```

### Renaming columns

```python
data=data.rename(columns={'NObeyesdad':'obesity_level'})
```

```python
data
```

## Preprocessing data
###  a) Checking missing values

```python
data.isnull().sum().transpose()
```

```python
data.duplicated().sum()
```

```python
duplicates=data[data.duplicated()]
print(duplicates.count())
```

```python
data.drop_duplicates(inplace=True)
data
```

```python
data=data.reset_index(level=None, drop=True, inplace=False, col_level=0, col_fill='')
```

```python
print(data.dtypes)
```

```python
from sklearn.preprocessing import LabelEncoder
data['Gender']=LabelEncoder().fit_transform(data['Gender'])
data['family_history_with_overweight']=LabelEncoder().fit_transform(data['family_history_with_overweight'])
data['FAVC']=LabelEncoder().fit_transform(data['FAVC'])
data['CAEC']=LabelEncoder().fit_transform(data['CAEC'])
data['SMOKE']=LabelEncoder().fit_transform(data['SMOKE'])
data['SCC']=LabelEncoder().fit_transform(data['SCC'])
data['CALC']=LabelEncoder().fit_transform(data['CALC'])
data['MTRANS']=LabelEncoder().fit_transform(data['MTRANS'])
data['obesity_level']=LabelEncoder().fit_transform(data['obesity_level'])
```

```python
data
```

```python
X=data.iloc[:,:-1]
```

```python
y=data.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
ranclf=RandomForestClassifier(criterion="gini", max_depth=100,min_samples_split=10,random_state=10)
ranclf.fit(X_train,y_train)
ranclf.score(X_test,y_test)
```

```python
import matplotlib.pyplot as plt
features=(ranclf.feature_importances_)*100
features
features=pd.DataFrame(features)
features['features_importance']=features
del features[0]
features_name=list(data.columns)
features_name.pop()
features['features']=features_name
features
```

```python
plt.figure()
features.plot.bar(x='features',y='features_importance',figsize=(10,5),color='green')
plt.xlabel("Data_columns",fontdict=dict(size=15))
plt.ylabel("% of importance of Data_columns",fontdict=dict(size=15))
plt.title("Data_columns vs % of importance",fontdict=dict(size=15))
plt.show()
```

```python
y_pred=ranclf.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(ranclf,X_train,y_train,cv=10)
```

```python
from sklearn.metrics import classification_report
report=classification_report(y_pred,y_test,output_dict=True)
print(report)
```

```python
each=pd.DataFrame(report).transpose()
new_report=each.drop([each.index[7],each.index[8],each.index[9]])
new_report
```

```python
names=['Insufficient_Weight','Normal_Weight','Obesity_Type_I','Obesity_Type_II','Obesity_Type_III','Overweight_Level_I',
       'Overweight_Level_II']
new_report['obesity_level']=names
new_report['precision']=new_report['precision']*100
new_report
```

```python
plt.figure()
new_report.plot.bar(x='obesity_level',y='precision',color='green',figsize=(10,5),rot=75)
plt.xlabel("obesity level",fontdict=dict(size=15))
plt.ylabel("% of Classifiaction accuracy",fontdict=dict(size=15))
plt.title("obesity vs % of Classification accuracy",fontdict=dict(size=15))
plt.show()
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
modeltree=DecisionTreeClassifier(max_leaf_nodes=8)
modeltree.fit(X_train,y_train)
modeltree.score(X_test,y_test)
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
names = ["insufficient","normal weight","obesity typeI","obesity type II","obesity type III","overweight level 1","overweight level 2"]

plt.figure(figsize=(20,10))
plot_tree(modeltree, feature_names = X.columns,class_names=names,rounded = True, filled= True)
plt.show()

```
