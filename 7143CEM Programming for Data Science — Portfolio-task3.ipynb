```python
#Part-1
# By using Matplotlib library
#Bar chart
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
IPL=pd.read_csv('IPLAuction.csv')
print(IPL)
AB=IPL.groupby(['Role','Year'])['Amount'].sum()
AC=AB.reset_index()
print(AC)
AD=AC.pivot(index='Year',columns='Role',values='Amount')
plt.figure()
AD.plot.bar(rot=75,figsize=(20,10))
plt.legend(bbox_to_anchor=(1,1))
plt.title('Action Amount vs Role of Player',fontdict= dict(size=30))
plt.ylabel('Amount (cores)',fontdict=dict(size=15))
plt.xlabel('Year',fontdict=dict(size=15))
plt.grid()
plt.show()
```

```python
# Bubbel plot

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
IPL=pd.read_csv('IPLAuction.csv')
print(IPL)
AB=IPL.groupby(['Role','Year'])['Amount'].sum()
AC=AB.reset_index()
AC['Role']=pd.Categorical(AC['Role'])
Role_list=AC.Role.unique()
print(Role_list)
plt.figure()
# fig,ax=plt.subplots(figsize=(20,10))
plt.figure(figsize=(20,10))
plt.scatter(x=AC['Year'],y=AC['Amount'],s=AC['Amount'],c=AC['Role'].cat.codes,
            cmap='Accent',alpha=0.5,sizes=(100,500),label=Role_list)
plt.title('Action Amount vs Role of Player',fontdict= dict(size=30))
plt.ylabel('Amount (cores)',fontdict=dict(size=15))
plt.xlabel('Year',fontdict=dict(size=15))
plt.grid()
plt.show()
```

```python
# Bar chart by using Seaborn Library
import seaborn as sns
import matplotlib.pyplot as plt
# print(seadata)
IPL=pd.read_csv('IPLAuction.csv')
print(IPL)
AB=IPL.groupby(['Role','Year'])['Amount'].sum()
AC=AB.reset_index()
print(AC)
plt.figure()
plt.figure(figsize=(20,10))
sns.barplot(x='Year',y='Amount',data=AC,hue='Role')
plt.legend(bbox_to_anchor=(1,1))
plt.title('Action Amount vs Role of Player',fontdict= dict(size=30))
plt.ylabel('Amount (cores)',fontdict=dict(size=15))
plt.xlabel('Year',fontdict=dict(size=15))
plt.xticks(rotation=60)
plt.grid()
plt.show()
```

```python
# Bubble chart by using seaborn 
import seaborn as sns
import matplotlib.pyplot as plt
# print(seadata)
IPL=pd.read_csv('IPLAuction.csv')
print(IPL)
AB=IPL.groupby(['Team','Year'])['Amount'].sum()
AC=AB.reset_index()
print(AC)
plt.figure(figsize=(20,10))
# plt.subplots(figsize=(20,10))
sns.scatterplot(x='Year',y='Amount',data=AC,hue='Team',size='Amount',sizes=(100,500))
plt.legend(bbox_to_anchor=(1,1))
plt.title('Action Amount vs Team',fontdict= dict(size=30))
plt.ylabel('Amount (cores)',fontdict=dict(size=15))
plt.xlabel('Year',fontdict=dict(size=15))
plt.xticks(rotation=60)
plt.grid()
plt.show()
```

```python
import plotly.express as px
import matplotlib.pyplot as plt
import pandas as pd
IPL=pd.read_csv('IPLAuction.csv')
print(IPL)
AB=IPL.groupby(['Role','Year'])['Amount'].sum()
AC=AB.reset_index()
fig=px.bar(x=AC['Year'],y=AC['Amount'],color=AC['Role'],title='Action Amount vs Role of Player',
           barmode='group',text=AC['Role'])
fig.show()
```

```python
import plotly.express as px
import matplotlib.pyplot as plt
import pandas as pd
IPL=pd.read_csv('IPLAuction.csv')
print(IPL)
AB=IPL.groupby(['Role','Year'])['Amount'].sum()
AC=AB.reset_index()
fig=px.scatter(x=AC['Year'],y=AC['Amount'],color=AC['Role'],
               title='Action Amount vs Role of Player',size=AC['Amount'],size_max=25)
fig.show()
```
