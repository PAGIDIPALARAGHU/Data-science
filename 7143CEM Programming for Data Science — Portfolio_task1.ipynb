```python
List=[1,2,8,96,5,6,7]
print("Length of list:",len(List))
def descendingorder(List):               
    count=0
    swap = True
    while (swap):
        swap = False
        for i in range(0,len(List)-1):       
            if (List[i] <= List[i+1]):
                count +=1
                print("Iteration No:",count)
                print("Before swap List=",List)
                print("i=",i,"List[i]=",List[i],"List[i+1]=",List[i+1])
                print("List[i] <= List[i+1] is",List[i]<=List[i+1])
                print("swap values of List[i],List[i+1]")
                a = List[i]
                print("Temporay value of a=",a)
                List[i]=List[i+1] 
                print("After swap value of List[i]=",List[i])
                List[i+1]=a 
                print("After swap value of List[i+1]=",a)
                swap = True
                print("After swap List=",List)   
    return(List) 
print("Final sorted List (Descending Order):",descendingorder(List))
```
