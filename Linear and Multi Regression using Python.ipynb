```python
import numpy as np
import matplotlib.pyplot as plt
X=np.array([[1,0],[1,3],[1,6]])
print("X:\n",X)
y=np.array([[1],[4],[5]])
print("y:\n",y)
XT=X.T
print("XT:\n",XT)
XTX=XT@X
print("XTX:\n",XTX)
invXTX=np.linalg.inv(XTX)
print("invXTX:\n",invXTX)
XTy=XT@y
print("XTy:\n",XTy)
β=invXTX@XTy
print("Matrix of β:\n",β)
print("intercept c:\n",β[0])
print("Multiplier m:\n",β[1])
Xβ=X@β
print("Xβ Matrix:\n",Xβ)
x=X[:,1]

#Fitted Graph

plt.figure(figsize=(8,6))
plt.scatter(x,y,c="red")
plt.plot(x,Xβ,'-o',c='green')
plt.title("Fitted Graph")
plt.xlabel("x",size=15)
plt.ylabel("Xβ",size=15)
plt.xlim(-1,8)
plt.ylim(-1,8)
plt.show()

#Finding value of R_Square
Er=y-Xβ
print("value of Error (y-Xβ):\n",Er)
SSEr=Er[0]**2+Er[1]**2+Er[2]**2
Mean_y=np.mean(y)
Er2=y-Mean_y
SSEr2=Er2[0]**2+Er2[1]**2+Er2[2]**2
R_square=1-(SSEr/SSEr2)
print("R_Square value is:",R_square)

# comparing Fitted Graph with polyfit Graph
x=X[:,1]
m,c=np.polyfit(x,y,1)
print("Intercept c:", c)
print("slope m:",m)
Y=m*x+c

plt.figure(figsize=(8,6))
plt.scatter(x,y,c="red")
plt.plot(x,Y,'-o',c='m')
plt.title("Fitted Graph with Polyfit")
plt.xlabel("x",size=15)
plt.ylabel("Y",size=15)
plt.xlim(-1,8)
plt.ylim(-1,8)
plt.show()



# Residual vs fitted Graph

x1=[0,7]
y1=[0,0]
plt.figure(figsize=(8,6))
plt.scatter(Xβ,Error,c='green')
plt.plot(x1,y1,'r--')
plt.title("Residual vs fitted")
plt.xlabel("Fitted value",size=15)
plt.ylabel("Residual(y-Xβ)",size=15)
plt.xlim(0,7)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

X=np.array([[1,0,0],[1,3,9],[1,6,36]])
print("X:\n",X)
y=np.array([[1],[4],[5]])
print("y:\n",y)
XT=X.T
print("XT:\n",XT)
XTX=XT@X
print("XTX:\n",XTX)
invXTX=np.linalg.inv(XTX)
print("invXTX:\n",invXTX)
XTy=XT@y
print("XTy:\n",XTy)
β=invXTX@XTy
print("Matrix of β:\n",β)
print("intercept c:\n",β[0])
print("coefficient of x:\n",β[1])
print("coefficient of x**2:\n",β[2])

#Fitted Graph
x=np.arange(0,6,0.1)
Y=β[2]*(x**2)+β[1]*x+β[0]
plt.figure(figsize=(8,6))
plt.plot(x,Y)
plt.title("Quadratic Graph ")
plt.xlabel("x",size=15)
plt.ylabel("Y",size=15)
plt.show()

#compare quadratic equation with polyfit
x=np.arange(0,6,0.1)
X1=X[:,1]
a,b,c=np.polyfit(X1,y,2)
print("Y1=a*(x**2)+b*x+c")
print("intercept x**2:\n",a)
print("intercept x:\n",b)
print("intercept c:\n",c)
Y1=a*(x**2)+b*x+c
plt.figure(figsize=(8,6))
plt.plot(x,Y1)
plt.title("Quadratic Graph with ployfit")
plt.xlabel("x",size=15)
plt.ylabel("Y1",size=15)
plt.show()
```

```python
import pandas as pd
coventry_day_length=pd.read_csv("coventry_day_length.csv")
print(coventry_day_length)
DL=coventry_day_length['Day length']
DL_Min=[]
for i in range(0,28):
    DLS=DL[i].split(':')
    DLM=int(int(DLS[0])*60+int(DLS[1])+int(DLS[2])/60)
    DL_Min.append(DLM)
print("Day length in Min:\n",DL_Min)
DLy=np.array([DL_Min])
D=coventry_day_length['Day No']
x=list(D)
print("Day No x:\n",x)
```
