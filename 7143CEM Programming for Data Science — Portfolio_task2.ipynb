```python
horse=input()
import random
def roll_8_face_dice(horse):
    body_count=0
    head_count=0
    leg_count=0
    tail_count=0
    eye_count=0
    mouth_count=0
    horn_count=0
    wing_count=0
    
    while (body_count<1):
        dice=random.randint(1,8)
        if dice==1:
            body_count +=1
            result=dict([(1,"body")])
            print("collect body")

    if horse=='unicorn':
        while(head_count<1):
            dice=random.randint(1,8)
            if dice!=1:
                if dice==2:
                    tail_count +=1
                    result.update([(2,"tail")])
                    print("collect tail and assemble to body")
                    
                if dice==3:
                    leg_count +=1
                    if leg_count<5:
                        result[3,leg_count]='leg'
                        print("collect leg and assemble to body")
                
                if dice==4:
                    head_count +=1
                    result.update([(4,"head")])
                    print("collect head and assemble to body")
                
        while(leg_count<4 or tail_count<1 or mouth_count<1 or
              eye_count<2 or horn_count<1):
            
            dice=random.randint(1,8)       
            if dice!=1 and dice!=4:
                if dice==2:
                    tail_count +=1
                    if tail_count<2:
                        result.update([(2,"tail")])
                        print("collect tail and assemble to body")
                    
                if dice==3:
                    leg_count +=1
                    if leg_count<5:
                        result[3,leg_count]='leg'
                        print("collect leg and assemble to body")
                        
                if dice==5:
                    eye_count +=1
                    if eye_count<3:
                        result[5,eye_count]='eye'
                        print("collect eye and assemble to head")
                    
                if dice==6:
                    mouth_count +=1
                    if mouth_count<2:
                        result.update([(6,"mouth")])
                        print("collect mouth and assemble to head")
                
                if dice==8:
                    horn_count +=1
                    if horn_count<2:
                        result.update([(8,"horn")])
                        print("collect horn and assemble to head")
    else:
        while(head_count<1):
            dice=random.randint(1,8)
            if dice!=1:
                if dice==2:
                    tail_count +=1
                    result.update([(2,"tail")])
                    print("collect tail and assemble to body")
                    
                if dice==3:
                    leg_count +=1
                    if leg_count<5:
                        result[3,leg_count]='leg'
                        print("collect leg and assemble to body")
                    
                if dice==7:
                    wing_count +=1
                    if wing_count<3:
                        result[7,wing_count]='wing'
                        print("collect wing and assemble to body")
                if dice==4:
                    head_count +=1
                    result.update([(4,"head")])
                    print("collect head and assemble to body")
                
        while(leg_count<4 or tail_count<1 or mouth_count<1 or 
              eye_count<2 or wing_count<2):
            
            dice=random.randint(1,8)        
            if dice!=1 and dice!=4:
                if dice==2:
                    tail_count +=1
                    if tail_count<2:
                        result.update([(2,"tail")])
                        print("collect tail and assemble to body")
                    
                if dice==3:
                    leg_count +=1
                    if leg_count<5:
                        result[3,leg_count]='leg'
                        print("collect leg and assemble to body")
                        
                if dice==5:
                    eye_count +=1
                    if eye_count<3:
                        result[5,eye_count]='eye'
                        print("collect eye and assemble to head")
                
                if dice==6:
                    mouth_count +=1
                    if mouth_count<2:
                        result.update([(6,"mouth")])
                        print("collect mouth and assemble to head")
                
                if dice==7:
                    wing_count +=1
                    if wing_count<3:
                        result[7,wing_count]='wing'
                        print("collect wing and assemble to body")
    return(result)
print(roll_8_face_dice(horse))
```
