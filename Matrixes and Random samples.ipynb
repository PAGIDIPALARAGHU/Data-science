```python
import numpy as np
import matplotlib.pyplot as plt
Q=np.array([[2/3,-2/3,1/3],[1/3,2/3,2/3],[2/3,1/3,-2/3]])
print("Given Matrix:\n",Q)
T=Q.T
print("Transpose of matrix:\n",Q@T)
I=np.identity(3)
print("Identity matrix:",I)
if np.allclose(Q@T,I)==True:
    print("Given Matrix Q is Orthogonal Matrix")
    C1=Q[:,0]
    C2=Q[:,1]
    C3=Q[:,2]
    print("First colum:\n",C1)
    print("Second column:\n",C2)
    print("Third column:\n",C3)
    MagC1=int(np.sqrt(C1[0]**2+C1[1]**2+C1[2]**2))                                   
    if MagC1==1:
        print("Magnitude of C1:",MagC1, "It is unit vector")
    else:
        print("First colum of given Matrix is not unit vector")
    MagC2=int(np.sqrt(C2[0]**2+C2[1]**2+C2[2]**2))
    if MagC2==1:
        print("Magnitude of C2:",MagC2, "It is unit vector")
    else:
        print("Second colum of given Matrix is not unit vector")
    MagC3=int(np.sqrt(C3[0]**2+C3[1]**2+C3[2]**2))
    if MagC3==1:
        print("Magnitude of C3:",MagC3, "It is unit vector")
    else:
        print("Third colum of given Matrix is not unit vector")
    P1=np.dot(C1,C2)
    
    print("P1 is dot product of C1,C2:",P1)
    MagP1=MagC1*MagC2
    P1_angle=int(((np.arccos(P1/MagP1))/(2*np.pi))*360)
    if P1_angle==90:
        print("Pair of C1 & C2 are orthogonal vectors")
        print("Angle between pair of C1 & C2:\n",P1_angle)
    else:
        print("Pair of C1 & C2 are not orthogonal vectors")
    P2=np.dot(C2,C3)
    print("P2 is dot product of C2,C3:",P2)
    MagP2=MagC2*MagC3
    P2_angle=int(((np.arccos(P2/MagP2))/(2*np.pi))*360)
    if P2_angle==90:
        print("Pair of C2 & C3 are orthogonal vectors")
        print("Angle between pair of C2 & C3:\n",P2_angle)
    else:
        print("Pair of C2 & C3 are not orthogonal vectors")
    P3=np.dot(C1,C3)
    print("P3 is dot product of C1,C3:",P3)
    MagP3=MagC1*MagC3
    P3_angle=int(((np.arccos(P3/MagP3))/(2*np.pi))*360)
    if P3_angle==90:
        print("Pair of C1 & C3 are orthogonal vectors")
        print("Angle between pair of C1 & C3:\n",P3_angle)
    else:
        print("Pair of C1 & C3 are not orthogonal vectors")
    T1=T[:,0]
    T2=T[:,1]
    T3=T[:,2]
    fig=plt.figure(figsize=(20,20))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot([0,C1[0]],[0,C1[1]],[0,C1[2]],'g',linewidth=3,label="Vector C1")
    ax.plot([0,C2[0]],[0,C2[1]],[0,C2[2]],'r',linewidth=3,label="Vector C2")
    ax.plot([0,C3[0]],[0,C3[1]],[0,C3[2]],'b',linewidth=3,label="Vector C3")
    ax.plot([0,T1[0]],[0,T1[1]],[0,T1[2]],'b:',linewidth=3,label="Vector T1")
    ax.plot([0,T2[0]],[0,T2[1]],[0,T2[2]],'r:',linewidth=3,label="Vector T2")
    ax.plot([0,T3[0]],[0,T3[1]],[0,T3[2]],'g:',linewidth=3,label="Vector T3")
    plt.xlabel("X",size=20)
    plt.ylabel("Y",size=20)
    plt.legend(loc="best")
    plt.show()
else:
    print("Given Matrix is not Orthogonal Matrix")
```

```python
#a)

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
print("Answer Part (a):")
mu=np.array([[9],[3]])
print("Mean  matrix shape:",mu.shape,"mu:\n",mu)
print("Mean of x is:",mu[0],"Mean of y is:",mu[1])
sigma=np.array([[2,5],[5,9]])
print("covaraince matrix always be a symmetric matrx")
print("Covariance matrix shape:",sigma.shape,"sigma:\n",sigma)
print("var(x):\n",sigma[0,:1],"\nvar(y):\n",sigma[1:,1:])
print("covariance(x,y) is:\n",sigma[0,1:])
n=10
sample = stats.multivariate_normal.rvs(size=n,
                                   mean=mu.flatten(),
                                   cov=sigma)
print("Random sample:\n",sample)
x=sample[:,0]
y=sample[:,1]
plt.figure(figsize=(15,10))
plt.scatter(x,y,c='green')
plt.xlabel("x",size=25)
plt.ylabel("y",size=25)
plt.title("Random sample",size=25)
plt.axis("equal")
plt.grid()
plt.show()
# calculating sample mean and sample covariance
sample_mux=np.mean(x)
print("sample mean of x:\n",sample_mux)
sample_muy=np.mean(y)
print("sample mean of y:\n",sample_muy)
C=np.cov(x,y)
print("Covariance Matrix C:\n",C)
print("sample mean values approxmently equal to mu x and y values")
print("sample covariance approxmently equal to assumend covariance")

print("======================================================================")

#b)
import numpy as np
print("Sample of random variable:",sample)
sample_mux=np.mean(x)
print("sample mean of x:\n",sample_mux)
sample_muy=np.mean(y)
print("sample mean of y:\n",sample_muy)
sample_mean=np.array([sample_mux,sample_muy])
print("sample mean matrix:\n",sample_mean)
Xc=sample-np.hstack(sample_mean)
print("Matrix of Xc1:\n",Xc)
C=np.cov(x,y)
print("Covariance Matrix C:\n",C)
XcT=Xc.T
print("Transpose of matrix XcT:\n",XcT)
XcTXc=XcT@Xc
print("Matrix of XcTXc:\n",XcTXc)
print("Relation between XcXcT and C is XcXcT=(n-1)*C")
XcTXc_R=(n-1)*C
print("Matrix XcTXc_R is:\n",XcTXc_R)
print("compare XcTXc_R & XcTXc :\n", np.isclose(XcTXc_R,XcTXc))
R= np.corrcoef(x,y)
print("Sample Correlation matrix R:\n",R)
DiaC=C.diagonal()
print("Diagonal elements of C:\n",DiaC)
sqrtDiaC=np.sqrt(DiaC)
print("Square root of diagonal elements of C:\n",sqrtDiaC)
E=np.diag(sqrtDiaC)
print("E is Square root of diagonal elements of C:\n",E)
invE=np.linalg.inv(E)
print("Inverese matrix E:\n",invE)
invET=invE.T
print("Transpose of the E:\n",invET)
print("R=invE@C@invET")
Rc=invE@C@invET
print("Rc Matrix:\n",Rc)
print("comparing R & Rc:\n",np.isclose(R,Rc))

print("======================================================================")
print("Answer Part (c):")
#c)
import numpy as np
C=np.cov(x,y)
print("covariance matrix:\n",C)
V,P=np.linalg.eig(C)
print("P matrix:\n", P)
D=np.diag(V)
print("D matrix:\n",D)
invP=np.linalg.inv(P)
print("Inverse matrix P:\n",invP)
PDinvP=P@D@invP
print("PDinvP:\n",PDinvP)
print(np.allclose(PDinvP,C))
print("Matrix of Xc1:\n",Xc)
print("Add P vector to Random Sample")
x=sample[:,0]
y=sample[:,1]
Px=P[:,0]
Py=P[:,1]
plt.figure(figsize=(15,10))
plt.scatter(x,y,c='green')
plt.scatter(Px,Py,c='red')
plt.xlabel("Px",size=25)
plt.ylabel("Py",size=25)
plt.title("Added P to Random sample",size=25)
plt.axis("equal")
plt.grid()
plt.show()
#New scatter plot of transformed matrix
PT=Xc@P
print("Transpformed matrix of centered sample:\n",PT)
PTx=PT[:,0]
PTy=PT[:,1]
plt.figure(figsize=(15,10))
plt.scatter(PTx,PTy,c='magenta')
plt.xlabel("PTX",size=25)
plt.ylabel("PTy",size=25)
plt.title("New scatter plot",size=25)
plt.axis("equal")
plt.grid()
plt.show()
PTC=np.cov(PTx,PTy)
print("Transformed centred sample covariance matrix PTC:\n",PTC)
print("D matrix:\n",D)
print(np.isclose(PTC,D))
print("D matrix and matrix PTC are similar:\n",np.allclose(PTC,D))
```
